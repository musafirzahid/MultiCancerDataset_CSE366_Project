```python


import os
import shutil
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input # IMPORT ADDED
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, confusion_matrix, 
                             roc_curve, auc)
from sklearn.preprocessing import label_binarize

# Set plot style
plt.style.use('seaborn-v0_8-whitegrid')

# --- 0. GPU CHECK ---
try:
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        print(f"✅ GPU Detected: {len(gpus)} device(s)")
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    else:
        print("⚠️ No GPU detected. Training will be slow.")
except RuntimeError as e:
    print(f"GPU Error: {e}")

# --- 1. CONFIGURATION & DIRECTORIES ---

DATASET_DIR = '/kaggle/input/cervical-cancer-largest-dataset-sipakmed'
WORKING_DIR = '/kaggle/working'
TRAIN_DIR = os.path.join(WORKING_DIR, 'train')
VAL_DIR = os.path.join(WORKING_DIR, 'val')
TEST_DIR = os.path.join(WORKING_DIR, 'test')

# Hyperparameters
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 15
LEARNING_RATE = 1e-4

# Clean up previous runs
if os.path.exists(TRAIN_DIR): shutil.rmtree(TRAIN_DIR)
if os.path.exists(VAL_DIR): shutil.rmtree(VAL_DIR)
if os.path.exists(TEST_DIR): shutil.rmtree(TEST_DIR)

# --- 2. DATASET SPLITTING ---

print("--- Preparing Dataset ---")
if not os.path.exists(DATASET_DIR):
    raise FileNotFoundError(f"Dataset not found at {DATASET_DIR}. Check your input paths.")

class_names = [d for d in os.listdir(DATASET_DIR) if os.path.isdir(os.path.join(DATASET_DIR, d))]
class_names.sort()

count_map = {}

for class_name in class_names:
    # Handle SIPaKMeD structure
    class_path = os.path.join(DATASET_DIR, class_name, class_name)
    if not os.path.isdir(class_path):
        class_path = os.path.join(DATASET_DIR, class_name)
    
    if not os.path.isdir(class_path): continue
        
    images = [os.path.join(class_path, img) for img in os.listdir(class_path) 
              if img.lower().endswith(('.bmp', '.jpg', '.png', '.jpeg'))]
    
    if not images: continue
    
    # Split: 70% Train, 15% Val, 15% Test
    train_imgs, temp_imgs = train_test_split(images, test_size=0.3, random_state=42, shuffle=True)
    val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.5, random_state=42, shuffle=True)
    
    count_map[class_name] = len(train_imgs) + len(val_imgs) + len(test_imgs)

    # Copy images
    for img_set, dest_dir in [(train_imgs, TRAIN_DIR), (val_imgs, VAL_DIR), (test_imgs, TEST_DIR)]:
        os.makedirs(os.path.join(dest_dir, class_name), exist_ok=True)
        for img in img_set:
            shutil.copy(img, os.path.join(dest_dir, class_name, os.path.basename(img)))

print(f"Dataset split complete. Found classes: {list(count_map.keys())}")

# --- 3. DATA GENERATORS (THE FIX) ---

# FIX: Removed rescale=1./255. Added preprocessing_function=preprocess_input
# ResNet expects inputs to be zero-centered, not 0-1 scaled.

train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input, # Critical for ResNet
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input # Critical for ResNet
)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR, target_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE, class_mode='categorical'
)

val_generator = val_test_datagen.flow_from_directory(
    VAL_DIR, target_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE, class_mode='categorical'
)

test_generator = val_test_datagen.flow_from_directory(
    TEST_DIR, target_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False
)

NUM_CLASSES = len(train_generator.class_indices)

# --- 4. MODEL BUILDING ---

def build_resnet50():
    print("\n--- Building ResNet50 Model ---")
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
    
    # Freeze base model
    base_model.trainable = False
    
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.5)(x)
    predictions = Dense(NUM_CLASSES, activation='softmax')(x)
    
    model = Model(inputs=base_model.input, outputs=predictions)
    
    model.compile(optimizer=Adam(learning_rate=LEARNING_RATE),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

model = build_resnet50()

# --- 5. TRAINING ---

print("\n--- Starting Training ---")
start_train_time = time.time()

# Added ReduceLROnPlateau to help convergence
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=[reduce_lr],
    verbose=1
)

total_train_time = time.time() - start_train_time
print(f"Training Time: {total_train_time:.2f} seconds")

# --- 6. PLOTTING GRAPHS ---

def plot_training_history_custom(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(1, len(acc) + 1)

    best_acc_epoch = np.argmax(val_acc)
    best_loss_epoch = np.argmin(val_loss)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    ax1.plot(epochs_range, loss, label='Training loss', color='indianred')
    ax1.plot(epochs_range, val_loss, label='Validation loss', color='mediumseagreen')
    ax1.scatter(best_loss_epoch + 1, val_loss[best_loss_epoch], color='blue', s=100, label=f'Best Epoch: {best_loss_epoch + 1}')
    ax1.set_title('Loss')
    ax1.legend()

    ax2.plot(epochs_range, acc, label='Training Accuracy', color='indianred')
    ax2.plot(epochs_range, val_acc, label='Validation Accuracy', color='mediumseagreen')
    ax2.scatter(best_acc_epoch + 1, val_acc[best_acc_epoch], color='blue', s=100, label=f'Best Epoch: {best_acc_epoch + 1}')
    ax2.set_title('Accuracy')
    ax2.legend()
    plt.show()

plot_training_history_custom(history)

# --- 7. EVALUATION ---

print("\n--- Model Evaluation ---")
y_pred_probs = model.predict(test_generator)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = test_generator.classes

# Generate correct class labels list
map_labels = {v: k for k, v in test_generator.class_indices.items()}
target_names = [map_labels[i] for i in range(NUM_CLASSES)]

# Classification Report
print(classification_report(y_true, y_pred, target_names=target_names))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
# --- 7. EVALUATION ---

print("\n--- Model Evaluation ---")
y_pred_probs = model.predict(test_generator)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = test_generator.classes

# Get correct class names in order
map_labels = {v: k for k, v in test_generator.class_indices.items()}
target_names = [map_labels[i] for i in range(len(map_labels))]

# Generate report in two formats: String (for the table) and Dict (for the custom list)
report_str = classification_report(y_true, y_pred, target_names=target_names)
report_dict = classification_report(y_true, y_pred, target_names=target_names, output_dict=True)

# 1. Print the Standard Classification Report
print("Classification Report:")
print(report_str)

# 2. Print the Custom Per-Class Accuracy (Recall) List
print("Per-Class Accuracy (Recall):")
for name in target_names:
    # Get the recall score, multiply by 100, and format to 2 decimal places
    recall_val = report_dict[name]['recall'] * 100
    print(f" - {name}: {recall_val:.2f}%")

# Optional: Confusion Matrix Plot
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
