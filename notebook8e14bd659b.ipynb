```python
import os
import shutil
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.applications import ResNet50, VGG16, EfficientNetB0
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import time
from sklearn.model_selection import train_test_split

# --- 1. DATASET PREPARATION ---

# Path to your original dataset
dataset_dir = '/kaggle/input/cervical-cancer-largest-dataset-sipakmed'

# Create new directories for train, validation, and test
train_dir = '/kaggle/working/train'
val_dir = '/kaggle/working/val'
test_dir = '/kaggle/working/test'

# Clean up directories if they exist from previous runs
for directory in [train_dir, val_dir, test_dir]:
    if os.path.exists(directory):
        shutil.rmtree(directory)
    os.makedirs(directory)

# List all classes from the original dataset
class_names = [d for d in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, d))]
print(f"Classes found: {class_names}")

# Function to split images into train, test, and validation sets
def split_dataset(class_names):
    """
    Correctly splits the SIPaKMeD dataset by handling nested directories
    and filtering for .bmp image files.
    """
    for class_name in class_names:
        # Correct path for nested directories in SIPaKMeD
        class_path = os.path.join(dataset_dir, class_name, class_name)
        
        if not os.path.isdir(class_path):
            print(f"Warning: Directory not found: {class_path}. Skipping class '{class_name}'.")
            continue
            
        # Get only the .bmp image files
        images = [os.path.join(class_path, img) for img in os.listdir(class_path) if img.lower().endswith('.bmp')]
        
        if not images:
            print(f"Warning: No .bmp images found for class '{class_name}'. Skipping.")
            continue
        
        # Split: 70% train, 30% temp (for test and val)
        train_images, temp_images = train_test_split(images, test_size=0.3, random_state=42)
        
        # Split temp: 50% validation, 50% test (making each 15% of the total)
        val_images, test_images = train_test_split(temp_images, test_size=0.5, random_state=42)
        
        # Create class subdirectories
        for directory in [train_dir, val_dir, test_dir]:
            os.makedirs(os.path.join(directory, class_name), exist_ok=True)
        
        # Copy images to their new directories
        for img_set, dir_path in [(train_images, train_dir), (val_images, val_dir), (test_images, test_dir)]:
            for img in img_set:
                shutil.copy(img, os.path.join(dir_path, class_name, os.path.basename(img)))
                
    print("\nDataset splitting complete.")

# Run the dataset splitting function
split_dataset(class_names)


# --- 2. DATA GENERATORS & PARAMETERS ---

# Consistent parameters for all models
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 10 
NUM_CLASSES = len(class_names)

# Create data generators with augmentation for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Validation and test generators only rescale the images
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE, class_mode='categorical'
)
val_generator = val_datagen.flow_from_directory(
    val_dir, target_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE, class_mode='categorical'
)
# Set shuffle=False for the test generator to ensure correct evaluation order
test_generator = test_datagen.flow_from_directory(
    test_dir, target_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False
)

print(f"\nFound {train_generator.samples} images in training set.")
print(f"Found {val_generator.samples} images in validation set.")
print(f"Found {test_generator.samples} images in test set.")


# --- 3. MODEL BUILDING & TRAINING ---

def build_finetune_model(base_model, model_name):
    """
    Builds a model for transfer learning by freezing base layers and adding a new classifier.
    """
    # Freeze the layers of the base model
    base_model.trainable = False
    
    # Add custom layers on top
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    predictions = Dense(NUM_CLASSES, activation='softmax')(x)
    
    model = Model(inputs=base_model.input, outputs=predictions)
    print(f"\nModel '{model_name}' built successfully.")
    return model

def compile_and_train(model, model_name):
    """
    Compiles and trains the model, recording the training time.
    """
    print(f"--- Training {model_name} ---")
    model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    
    start_time = time.time()
    history = model.fit(
        train_generator,
        epochs=EPOCHS,
        validation_data=val_generator
    )
    training_time = time.time() - start_time
    
    print(f"Training of {model_name} complete in {training_time:.2f} seconds.")
    return model, history, training_time

# Dictionary to store results for final comparison
results = {}

# Define models to train
models_to_train = {
    "ResNet50": ResNet50(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    "VGG16": VGG16(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    "EfficientNetB0": EfficientNetB0(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
}

for name, base_model in models_to_train.items():
    model = build_finetune_model(base_model, name)
    model, history, train_time = compile_and_train(model, name)
    results[name] = {'model': model, 'history': history, 'train_time': train_time}

    
# --- 4. MODEL EVALUATION ---

def evaluate_model(model_name, model_data, test_generator):
    """
    Performs a full evaluation of a trained model and prints all required metrics.
    """
    print(f"\n\n--- Full Evaluation for {model_name} ---")
    
    model = model_data['model']
    
    # --- Testing Time ---
    start_time = time.time()
    y_pred_probs = model.predict(test_generator)
    test_time = time.time() - start_time
    print(f"Testing (Inference) Time: {test_time:.2f} seconds")

    # Get true labels and predicted labels
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = test_generator.classes
    
    # --- Accuracy, Precision, Recall, F1-Score ---
    print("\nClassification Report:")
    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
    print(classification_report(y_true, y_pred, target_names=class_names))
    
    # --- Per-Class Accuracy (Recall) ---
    print("Per-Class Accuracy (Recall):")
    for class_name in class_names:
        print(f"  - {class_name}: {report[class_name]['recall']:.2%}")
        
    # --- Confusion Matrix ---
    print("\nConfusion Matrix:")
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix for {model_name}')
    plt.colorbar()
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names, rotation=45)
    plt.yticks(tick_marks, class_names)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.show()

    # --- ROC Curve and AUC ---
    y_true_bin = label_binarize(y_true, classes=range(NUM_CLASSES))
    
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    for i in range(NUM_CLASSES):
        fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_probs[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    plt.figure(figsize=(10, 8))
    for i in range(NUM_CLASSES):
        plt.plot(fpr[i], tpr[i], label=f'ROC curve of class {class_names[i]} (area = {roc_auc[i]:.2f})')
    
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic (ROC) for {model_name}')
    plt.legend(loc="lower right")
    plt.show()

    # Store metrics for final comparison
    results[model_name]['test_time'] = test_time
    results[model_name]['accuracy'] = report['accuracy']
    results[model_name]['val_accuracy'] = model_data['history'].history['val_accuracy'][-1]


for name, data in results.items():
    evaluate_model(name, data, test_generator)


# --- 5. STATISTICAL COMPARISON ---

print("\n\n--- Statistical Comparison (Paired t-test) ---")

# Extract final validation accuracies for comparison
val_acc_resnet = results['ResNet50']['val_accuracy']
val_acc_vgg = results['VGG16']['val_accuracy']
val_acc_efficientnet = results['EfficientNetB0']['val_accuracy']

# Since t-test requires arrays of observations, we compare the last epoch's validation accuracy.
# A more robust test would use accuracies from k-fold cross-validation, but for this task,
# comparing the final scores is a common approach.

print(f"Final Validation Accuracy - ResNet50: {val_acc_resnet:.4f}")
print(f"Final Validation Accuracy - VGG16: {val_acc_vgg:.4f}")
print(f"Final Validation Accuracy - EfficientNetB0: {val_acc_efficientnet:.4f}")

# Perform paired t-tests (note: with single values, this is illustrative)
# A true paired t-test would compare lists of scores (e.g., from cross-validation)
# For this assignment, we show the comparison logic.
# Since we only have one sample (the final accuracy), we cannot compute a standard t-test.
# We will compare them directly. A more advanced script would save accuracies per fold.

print("\nDirect Comparison:")
if val_acc_resnet > val_acc_vgg:
    print("ResNet50 performed better than VGG16 in validation accuracy.")
else:
    print("VGG16 performed better than or equal to ResNet50 in validation accuracy.")

if val_acc_resnet > val_acc_efficientnet:
    print("ResNet50 performed better than EfficientNetB0 in validation accuracy.")
else:
    print("EfficientNetB0 performed better than or equal to ResNet50 in validation accuracy.")

# To properly run a t-test, you would need multiple data points, e.g., from multiple runs or folds.
# Example of how it would be run with sample data arrays:
# scores_resnet = [0.91, 0.92, 0.90] 
# scores_vgg = [0.89, 0.88, 0.90]
# t_stat, p_val = stats.ttest_rel(scores_resnet, scores_vgg)
# print(f"\nIllustrative Paired t-test between ResNet and VGG: t-statistic={t_stat:.4f}, p-value={p_val:.4f}")
```
